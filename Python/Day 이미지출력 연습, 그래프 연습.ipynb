# matplot 를 이용하여 그래프 그리기



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
y = x ** 2
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(10)
y = x ** 2
plt.title("y=x**2")   #위에 제목
plt.xlabel('x')    #x축에 x라고 표시
plt.ylabel('y') #y축에 y라고 표시
plt.grid(True)  #격자 무늬
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(10)
y = x ** 2
plt.title("y=x**2")   #위에 제목
plt.xlabel('x')    #x축에 x라고 표시
plt.ylabel('y') #y축에 y라고 표시
plt.grid(True)  #격자 무늬
plt.plot(x,y, label ='y=x**2')

plt.plot(x*2, label ='y=x*2')  #그래프에 이름을 붙일때는 레이블이 있어야한다. 
plt.legend()  #그래프에 이름 붙이기


plt.show()
```

```python
x = np.arange(10)
y = x ** 2
plt.title("y=x**2")   #위에 제목
plt.xlabel('x')    #x축에 x라고 표시
plt.ylabel('y') #y축에 y라고 표시
plt.grid(True)  #격자 무늬
plt.plot(x,y, label ='y=x**2', color ='red')

plt.plot(x*2, label ='y=x*2', color ='black')  #그래프에 이름을 붙일때는 레이블이 있어야한다. 
plt.legend()  #그래프에 이름 붙이기


plt.show()
```

```python
x= np.arange(10)
y = x **2 

plt.plot(x,y, label ='y=x**2', color ='red')
plt.show()

plt.plot(x*2, label ='y=x*2', color ='black')
plt.show()
```

# 서브플롯 //// 그래프의 위치를 움직일수 있다. 


```python
x= np.arange(10)
y = x **2 
#subplot(행, 열 , 인덱스 )
plt.subplot(2,3,1)
plt.plot(x,y, label ='y=x**2', color ='red')

plt.subplot(2,3,2)
plt.plot(x,y, label ='y=x**2', color ='red')

plt.subplot(2,3,3)
plt.plot(x,y, label ='y=x**2', color ='red')

plt.subplot(2,3,4)
plt.plot(x,y, label ='y=x**2', color ='red')

plt.subplot(2,3,5)
plt.plot(x,y, label ='y=x**2', color ='red')

plt.subplot(2,3,6)
plt.plot(x,y, label ='y=x**2', color ='red')

plt.show()
```

```python
plt.hist(np.random.rand(100000),bins=100)#균등분포
plt.show()

plt.hist(np.random.randn(100000),bins=100)#노멀분포(정규분포)우리가 사용할것
plt.show()
```
