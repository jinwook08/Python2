```python
def plot_KMeans(n):
    model = cluster.KMeans(n_clusters =2,n_init = 1, random_state = 42).fit(minmaxdata)
    c0,c1 = model.cluster_centers_
    plt.scatter(minmaxdata[model.labels_ ==0,0], minmaxdata[model.labels_ == 0,1], marker = 'v', facecolor = 'r', edgecolor = 'k')
    plt.scatter(minmaxdata[model.labels_ ==1,0], minmaxdata[model.labels_ == 1,1], marker = '^', facecolor = 'y', edgecolor = 'k')
    plt.scatter(c0[0], c0[1], marker='v', c="r", s=200)
    plt.scatter(c1[0], c1[1], marker='^', c="y", s=200)
    plt.grid(False)
    plt.title("iteration={}, score={:5.2f}".format(n, model.score(minmaxdata)))
    
    
plt.figure(figsize=(8, 8))
plt.subplot(321)
plot_KMeans(1)
plt.subplot(322)
plot_KMeans(2)
plt.subplot(323)
plot_KMeans(3)
plt.subplot(324)
plot_KMeans(4)
plt.tight_layout()
plt.show()
```
