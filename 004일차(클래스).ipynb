```python
# 사원관리

class Employee : 
    empCount = 0  # 사원수를 저장 : 클래스 변수 생성
    def __init__(self, name, salary) :
        self.name = name # 이름
        self.salary = salary # 급여
        Employee.empCount += 1
    def displayCount(self) :
        print("직원 수 = %d" % Employee.empCount)
    def displayEmployee(self) :
        print("이름 : ", self.name,  "급여 : ",self.salary)
    def __call__(self, *pargs, **kargs) :
        print( ' Called : ', pargs, kargs )
        
emp = Employee("현동엽", 500)
print(emp.displayCount())  # 1
emp1 = Employee("빅데이터", 600)
print(emp1.displayCount())  # 2
print(emp.displayCount())   # 2  => empCount가 공유변수이기 때문에 바뀜
print(emp.displayEmployee())   # 데이터를 담고있는 위치가 다르기 때문에 다른 값 
print(emp1.displayEmployee())
print(emp("빅데이터", 500, 100, c = 100)) # 클래스를 함수처럼 사용하고 싶을 때, __call__을 overring하면 됨
```

```python
import math

# x1 = 10, y1 = 20
# x2 = 20, y2 = 30
#distance = math.sqrt((x1-x2)**2 + (y1-y2)**2)

# 1개를 다루는 클래스를 작성하고 있지만, 거리값은 두 개를 이용
class Point :
    def __init__(self, x ,y ) :  # 생성자
        self.x=x
        self.y=y
    def __str__(self) :   # __reqr__ 문자열 요구하는 함수에 들어가면 작동
        return "좌표는 (%d , %d)" % (self.x, self.y)
    def dist(self, other ) :
        distance = math.sqrt((self.x- other.x)**2 + (self.y-other.y)**2)
        return distance
    def collision(self,other) :
        return self.dist(other) < 3
    
    
```

```python
first = Point(7,3)
second = Point(4,7)
print(first)
print(second)

print("두 점 사이의 거리는 = ", first.dist(second))
if first.collision(second) :
    print("충돌했습니다.")
else :
    print("충돌하지 않았습니다.")
```

```python
# 선생님 답

class circle :   # center = Point()
    def __init__(self,center, radius) :
        self.center = center
        self.radius = radius
    def __repr__(self) : #문자열 캐스팅
        return "원의 중심은 " + str(self.center) + "반지름은 " + str(self.radius)
    def collision(self, other) :
        if self.center.dist(other.center) < self.radius + other.radius :
            result = "충돌한다"
        else :
            return "충돌하지 않는다."
        return result

cir = circle(first, 1)
cir2 = circle(second, 5)
print(cir)
print(cir2)

print(cir.collision(cir2))
circle.collision(cir, cir2)  # self로  cir이 들어간 것
```

```python
# Point를 이용해서 사격형 Rectangle 클래스를 작성하시오
# - 사격형의 면적을 구하는 함수를 작성
# - 사각형의 둘레를 구하는 함수를 작성
# - 대각선의 길이를 구하는 함수를 작성

class rectangle :
    def __init__(self, point, point2) :  # point = Point()  / point2 = Point() 맵핑
        self.point = point
        self.point2 = point2
    def area(self) :
        return abs(self.point.x - self.point2.x) * abs(self.point.y - self.point2.y)
    def diagonal_dist(self) :
        return self.point.dist(self.point2)
    def perimeter(self) :
        return abs(self.point.x - self.point2.x) * 2 + abs(self.point.y - self.point2.y)* 2
    
distance = rectangle(first,second)
print(distance.area())
print(distance.diagonal_dist())
print(distance.perimeter())
```

```python
# 문제 : Vector Class를 작성하세요.
# x ,y 의 2차원 Vector 구현
# Vector의 연산 ( + , - , * , / )  : 요소별로 계산하면 됨
# dot함수를 구현하세요 : Vector의 요소별로 곱하고, 모두 더하기 ( 내적 유사도 )
# cos 값으로 변환하는 함수를 구현하세요.
#    cos theta = acos( A내적B / |A||B| )
#   |A| = math.sqrt( x**2 + y**2 )



# a = 10
# b = 20
# c = a + b


class Vector : 
    def __init__(self, x , y) :
        self.x = x
        self.y = y
    def __repr__(self) :
        return "(좌표는 %d, %d)"  % (self.x , self.y)
    
    def __add__(self,other) :                               # 연산자 오버로딩을 오버라이딩으로 구현한 것
        return Vector(self.x + other.x, self.y + other.y)
    # Vector(10, 20) + Vector(30 , 40)  ==> 그냥 더하기만 하면 어떤 것을 더하는지 모름 ==> +(__add__ 함수)를 재정의함
    
    def __mul__(self, other) :
        return Vector(self.x * other.x, self.y * other.y)
    def __sub__(self,other) :
        return Vector(self.x - other.x, self.y - other.y)
    def __truediv__(self,other) :  # div : python 2 버전에서 사용
        return Vector(self.x / other.x, self.y / other.y)
```

```python
# 문제 : 3차원 벡터를 클래스로 구현하시오( 연산자 오버로딩 +, - , * , / )

class Vector : 
    def __init__(self,x,y,z) :
        self.x= x
        self.y = y
        self.z = z
    def __repr__(self) :
        return "(좌표는 %d, %d, %d)" % (self.x,self.y,self.z)
    def __add__(self,other) :
        return Vector(self.x + other.x, self.y + other.y, self.z+ other.z)
    def __mul__(self, other) :
        return Vector(self.x * other.x, self.y * other.y, self.z * other.z)
    def __sub__(self,other) :
        return Vector(self.x - other.x, self.y - other.y, self.z - other.z)
    def __truediv__(self,other) :
        return Vector(self.x / other.x, self.y / other.y, self.z / otehr.z)
    def dist(self) :
        return math.sqrt(self.x*self.x + self.y * self.y + self.z* self.z)
    
    # 방향만 바뀌어야지 크기까지 바뀌면 안됨
    def norm(self) : # 정규화( normalization ) :  크기값이 1인 사이즈로 변경( 반지름이 1인 사이즈 )
        dist = self.dist()
        return self.x/dist, self.y/dist, self.z/dist
    def dot(self, other ) : # 내적 
        return (self.x*other.x + self.y*other.y + self.z*other.z)
    def theta(self, other) :
        dot = self.dot(other)
        dist1 = self.dist()
        dist2 = other.dist() 
        return math.acos(dot/(dist1*dist2)) * (180/math.pi)  #radian -> degree로 변경  ( 컴퓨터는 radian만 사용 )
# 변환과정이 필요 :: 사람이 사용하는 각도 입력 => 라디안 변환(math.pi/180) => 계산 후 (180/math.pi )
# => 사람이 사용하는 degree로 변환해서 출력
        
    
    
        
f = Vector(3,4,5)
g = Vector(6, 7, 8)
print(f)
print(f+g)
print(f-g)
```

```python
# 2개로  분리된 class
# - 기본적인 데이터를 저장하는 클래스
# - Management( 입력, 출력, 검색, 수정 등 )
# - Menu 함수


# 1인분에 대한 처리를 정의
import sys
class Student :
    def __init__(self) :
        self.num = 0
        self.name = ''
        self.kor = 0
        self.math = 0
        self.eng = 0
        self.total = 0
        self.avg = 0
        self.grade =""
    def inputData(self) :
        self.name = input("이름 : ")
        self.kor = int(input("국어 점수 : "))
        self.math = int(input("수학 점수 : "))
        self.eng = int(input("영어 점수 : "))
        
    def calc_total_avg(self) :    # 수정을 고려해서 별도의 함수화
        self.total = self.kor + self.math + self.eng
        self.avg = round( self.total / 3,2 )
        self.grade = self.calcSemGrade()
        
    def calcSemGrade(self) :
        if self.avg >= 90 :
            return "A"
        elif self.avg >= 80 :
            return "B"
        elif self.avg >= 70 :
            return "C"
        elif self.avg >= 60 :
            return "D"
        else :
            return "F"
        
    def __str__(self) :
        return '%5s %5s %6s %6s %6s %7s %7s %7s' % (self.num, self.name, self.kor, self.math,
                                                       self.eng,self.total,self.avg,self.grade)
    
    def __cmp__(self,other) : # compare  === True/False로 리턴
        return self.name == other.name
```

```python
# sj = ["번호", "이름","국어","수학","영어","총점","평균","학점"]

menu = ["입력(1), 출력(2), 계산(3), 검색(4), 이름수정(5), 이름삭제(6), 정렬(7), 종료(8)"]
man_sung = Management(1)

while 1 :
    sel = input(menu)
    if sel == "1" :
        man_sung.input()
        continue
    elif sel == "2" :
        print(Management.schoolname + " 성적 계산표 ")
        print("%5s %5s %6s %6s %6s %7s %7s %7s" % ("번 호", "이 름", "국 어","수 학","영 어","총 점","평 균","학 점"))
        man_sung.print_score()
        continue
    elif sel == "3" :
        man_sung.calc_score()
        print("계산이 완료 되어있습니다. - 확인하려면 출력을 선택하세요.")
        continue
    elif sel =="4" :
        man_sung.search_name()
        continue
    
    elif sel == "5" :
        man_sung.change_name()
        print("이름 변경이 완료되었습니다.")
        continue
              
    elif sel =="6" :
        man_sung.remove_name()
        print("이름 삭제가 완료되었습니다.")
        continue
    elif sel =="7" :
        man_sung.sort_total()
    elif sel =="8":
        break
    else :
        print("잘못된 입력입니다. 다시 입력하세요")
        continue
```
